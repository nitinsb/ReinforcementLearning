```python
import numpy as np
```

```python
S = ('s1','s2')
A = ('a1','a2')
# triple dictonaries
p = {s:{a:{s_next:0 for s_next in S} for a in A } for s in S}
r = {s:{a:{s_next:0 for s_next in S} for a in A } for s in S}
p['s1']['a1']['s1'] = 0.8
p['s1']['a1']['s2'] = 0.2
p['s1']['a2']['s1'] = 0.5
p['s1']['a2']['s2'] = 0.5
p['s2']['a1']['s1'] = 0.7
p['s2']['a1']['s2'] = 0.3
p['s2']['a2']['s1'] = 0.4
p['s2']['a2']['s2'] = 0.6

# expected reward

r['s1']['a1']['s1'] = 4
r['s1']['a1']['s2'] = 4
r['s1']['a2']['s1'] = 6
r['s1']['a2']['s2'] = 6
r['s2']['a1']['s1'] = -5
r['s2']['a1']['s2'] = -5
r['s2']['a2']['s1'] = -3
r['s2']['a2']['s2'] = -3

gamma = 0.9
```

```python
thetha = 1e-3
delta = thetha + 1
v = { s : 0 for s in S} # intitilisation
iter = 0
#one iteration of value iteration
while delta >= thetha:
    iter += 1
    q = {s:{a:0 for a in A} for s in S}
    v_next = {s:0 for s in S}
    delta = 0
    #one Sweep of State space
    for s in S:
        for a in A:
            for s_next in S:
                q[s][a] += p[s][a][s_next]*(r[s][a][s_next] + gamma*v[s_next])
                #v_next[s] += pi[s][a]*prob*(reward + gamma * v[s_next])
                # once we compute above we need to get max of it as in equation for all actions get max action
            #come out of action loop as well
        v_next[s] = max([q[s][a] for a in A])
        delta = max(delta, np.abs(v_next[s] - v[s]))
        
    v = v_next.copy()
    
print(iter , thetha)
```
